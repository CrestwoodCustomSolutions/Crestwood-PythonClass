```python
# Lesson 12.01
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["Maria", "Luka", "Chris", np.nan],
    "Age": [25,27,44,35],
    "Occupation": ["Engineer","Competitive Eater", "Artist", "Scientist"]
})

df.isna()
df.dropna(inplace=True)
df

df.fillna("0")
```

```python
# Lesson 12.02
import pandas as pd
from scipy import stats

df2 = pd.DataFrame({
    "Transaction_Amount": [100, 150, 200, 250, 300, 9999]
})

df2

df2['Z_Score'] = stats.zscore(df2['Transaction_Amount'])
df2['Winsorized_Amount'] = stats.mstats.winsorize(df2['Transaction_Amount'], limits=[0.05, 0.05])
df2

threashold = 2
df2['Outlier'] = df2['Z_Score'].abs() > threashold
df2

df2_truncate = df2[df2['Outlier'] == False]
df2_truncate
```
